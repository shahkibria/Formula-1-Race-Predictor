```python
# We will finish all the statuses where drivers finish the race to 2
replace = {
    11:2, 12:2, 13:2, 14:2, 15:2, 16:2, 17:2, 18:2, 19:2, 45:2, 50:2, 128:2, 53:2, 55:2, 58:2, 
    88:2, 111:2, 112:2, 113:2, 114:2, 115:2, 116:2, 117:2, 118:2, 119:2, 120:2, 122:2, 123:2, 124:2, 
    125:2, 127:2, 133:2, 134:2  
}
df["statusId"] = df["statusId"].replace(replace)
```

```python
# Transform finish_status column. Anything which is not 1 (driver did not finish race) is converted to 3
def change_status(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    else:
        return 3
    
df["statusId"] = df["statusId"].apply(change_status)
df.tail(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df.statusId.values.reshape(-1,1)))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(['statusId'])
encode_df.head()
```

```python
df = df.merge(encode_df,left_index=True,right_index=True).drop("statusId",1)
```

```python
# Transform finish_status column. Anything which is not 1 (driver did not finish race) is converted to 3
podium = [1,2,3]
def change_order(x):
    if x == 1:
        return 1
    else:
        return 0

def podium_order(x):
    if x in podium:
        return x
    else:
        return 0
df["podium"] = df["positionOrder"].apply(podium_order)    
df["positionOrder"] = df["positionOrder"].apply(change_order)

df.head(10)
```

```python
import sklearn as skl
from sklearn import preprocessing
X_scaler = skl.preprocessing.StandardScaler()
```

```python
df_train = df[(df['year_x']<=2018) & (df['year_x']>=2000)]
df_test = df[df['year_x']==2019]
```

```python
df_train
```

```python
y_train= df_train['positionOrder']
y_test= df_test['positionOrder']
X_train = df_train.drop(columns=['positionOrder','podium'])
# .drop(columns=['positionOrder','driverId','constructorId','circuitId','year_x'])
X_test = df_test.drop(columns=['positionOrder','podium'])
# X_test = df_test.drop(columns=['positionOrder','driverId','constructorId','circuitId','year_x'])
```

```python
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly',probability=True)
```

```python
# Fit the data
model.fit(X_train_scaled, y_train)
```

```python
predictions = model.predict_proba(X_test_scaled)
predictions
# test = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
# test
```

```python
pred_df=pd.DataFrame(predictions,columns=['p0','p1'])
```

```python
test = pd.merge(df_test.reset_index().drop(columns=['index']),pred_df,left_index=True,right_index=True)
# test = pd.merge(test,circuitId_df,left_index=True,right_index=True)
test
```

```python
max_prob = test.groupby(['circuitId']).agg({'p1':'max'})
test = pd.merge(test, max_prob, how='left', on=['circuitId'])
# test = pd.merge(test, podium_df, right_index=True,left_index=True)
test
```
