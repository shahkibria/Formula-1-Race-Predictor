```python
# We will finish all the statuses where drivers finish the race to 2
replace = {
    11:2, 12:2, 13:2, 14:2, 15:2, 16:2, 17:2, 18:2, 19:2, 45:2, 50:2, 128:2, 53:2, 55:2, 58:2, 
    88:2, 111:2, 112:2, 113:2, 114:2, 115:2, 116:2, 117:2, 118:2, 119:2, 120:2, 122:2, 123:2, 124:2, 
    125:2, 127:2, 133:2, 134:2  
}
df["statusId"] = df["statusId"].replace(replace)
```

```python
# Transform finish_status column. Anything which is not 1 (driver did not finish race) is converted to 3
def change_status(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    else:
        return 3
    
df["statusId"] = df["statusId"].apply(change_status)
df.tail(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df.statusId.values.reshape(-1,1)))

# Rename encoded columns
encode_df.columns = enc.get_feature_names(['statusId'])
encode_df.head()
```

```python
df = df.merge(encode_df,left_index=True,right_index=True).drop("statusId",1)
```

```python
# Transform finish_status column. Anything which is not 1 (driver did not finish race) is converted to 3
podium = [1,2,3]
def change_order(x):
    if x == 1:
        return 1
    else:
        return 0

def podium_order(x):
    if x in podium:
        return x
    else:
        return 0
df["podium"] = df["positionOrder"].apply(podium_order)    
df["positionOrder"] = df["positionOrder"].apply(change_order)

df.head(10)
```

```python
podiumdf = df[df['year_x']>=2019]['podium']
podiumdf
```

```python
# Fit the encoder and produce encoded DataFrame
encode_df = pd.DataFrame(enc.fit_transform(df.circuitId.values.reshape(-1,1)))
# Rename encoded columns
encode_df.columns = enc.get_feature_names(['circuitId'])
df = df.merge(encode_df,left_index=True,right_index=True).drop("circuitId",1)
```

```python
import sklearn as skl
from sklearn import preprocessing
X_scaler = skl.preprocessing.StandardScaler()
```

```python
df_train = df[(df['year_x']<=2018) & (df['year_x']>=2000)]
df_test = df[df['year_x']>=2019]
```

```python
df_train
```

```python
y_train= df_train['positionOrder']
y_test= df_test['positionOrder']
X_train = df_train.drop(columns=['positionOrder','podium'])
# .drop(columns=['positionOrder','driverId','constructorId','circuitId','year_x'])
X_test = df_test.drop(columns=['positionOrder','podium'])
# X_test = df_test.drop(columns=['positionOrder','driverId','constructorId','circuitId','year_x'])
```

```python
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
import tensorflow as tf
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()

# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=150, activation="tanh", input_dim=1147))
nn_model.add(tf.keras.layers.Dense(units=100, activation="tanh"))
nn_model.add(tf.keras.layers.Dense(units=50, activation="relu"))
# nn_model.add(tf.keras.layers.Dense(units=25, activation="relu"))
# nn_model.add(tf.keras.layers.Dense(units=10, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=10, activation="sigmoid"))
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
# Check the structure of the Sequential model
nn_model.summary()
```

```python
nn_model.compile(optimizer="SGD", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50)
```

```python
nn_model.evaluate(X_test_scaled,y_test,verbose=2)
```

```python
predictions = nn_model.predict(X_test_scaled)
```
