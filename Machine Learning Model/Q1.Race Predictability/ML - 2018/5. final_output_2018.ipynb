```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Read all four output files

classifier = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2018/cl_output_2018.csv'))
classifier = classifier.drop(['Unnamed: 0'], axis=1)

svm = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2018/svm_output_2018.csv'))
svm = svm.drop(['Unnamed: 0'], axis=1)

rnf = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2018/rnf_output_2018.csv'))
rnf = rnf.drop(['Unnamed: 0'], axis=1)

nn = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2018/nn_output_2018.csv'))
nn = nn.drop(['Unnamed: 0'], axis=1)
```

```python
# Merge all 4 files on result ID
output_2018 = pd.merge(classifier, svm, how='left', on=['resultId'])
```

```python
output_2018 = pd.merge(output_2018, rnf, how='left', on=['resultId'])
```

```python
output_2018 = pd.merge(output_2018, nn, how='left', on=['resultId'])
output_2018.head()
```

```python
# Extract CSV file
output_2018.to_csv('../../../Resources/PythonExport/ML_Export/2018/output_2018.csv')
```
