```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Read all four output files

classifier = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2017/cl_output_2017.csv'))
classifier = classifier.drop(['Unnamed: 0'], axis=1)

svm = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2017/svm_output_2017.csv'))
svm = svm.drop(['Unnamed: 0'], axis=1)

rnf = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2017/rnf_output_2017.csv'))
rnf = rnf.drop(['Unnamed: 0'], axis=1)

nn = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2017/nn_output_2017.csv'))
nn = nn.drop(['Unnamed: 0'], axis=1)
```

```python
# Merge all 4 files on result ID
output_2017 = pd.merge(classifier, svm, how='left', on=['resultId'])
```

```python
output_2017 = pd.merge(output_2017, rnf, how='left', on=['resultId'])
```

```python
output_2017 = pd.merge(output_2017, nn, how='left', on=['resultId'])
output_2017.head()
```

```python
# Extract CSV file
output_2017.to_csv('../../../Resources/PythonExport/ML_Export/2017/output_2017.csv')
```
