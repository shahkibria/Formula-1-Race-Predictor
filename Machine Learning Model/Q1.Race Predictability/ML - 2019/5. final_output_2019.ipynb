```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Read all four output files

classifier = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2019/cl_output_2019.csv'))
classifier = classifier.drop(['Unnamed: 0'], axis=1)

svm = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2019/svm_output_2019.csv'))
svm = svm.drop(['Unnamed: 0'], axis=1)

rnf = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2019/rnf_output_2019.csv'))
rnf = rnf.drop(['Unnamed: 0'], axis=1)

nn = pd.read_csv(Path('../../../Resources/PythonExport/ML_Export/2019/nn_output_2019.csv'))
nn = nn.drop(['Unnamed: 0'], axis=1)
```

```python
# Merge all 4 files on result ID
output_2019 = pd.merge(classifier, svm, how='left', on=['resultId'])
```

```python
output_2019 = pd.merge(output_2019, rnf, how='left', on=['resultId'])
```

```python
output_2019 = pd.merge(output_2019, nn, how='left', on=['resultId'])
output_2019.head()
```

```python
# Extract CSV file
output_2019.to_csv('../../../Resources/PythonExport/ML_Export/2019/output_2019.csv')
```
